```python
#ch03~06
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-3,3,50)
y1=2*x+1
y2=x**2

#num=圖表編號，figsize=圖表大小
plt.figure(num=3,figsize=(8,5))
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')

#調整可視範圍
plt.xlim((-1,2))
plt.ylim((-2,3))

#調整向量名
plt.xlabel("I am x")
plt.ylabel("I am y")

#調整刻度單位
new_ticks=np.linspace(-1,2,5)
plt.xticks(new_ticks)
plt.yticks([-2,-1.8,-1,1.22,3],
          ['really bad','bad','normal','good','really good'])

#坐標軸信息
ax=plt.gca()

#調整邊框顏色
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#調整刻度位置
ax.xaxis.set_ticks_position('bottom') #top，bottom，both，default，none
ax.yaxis.set_ticks_position('left')  #left，right，both，default，none

#調整坐標軸位置
ax.spines['bottom'].set_position(('data',0)) #讓x軸對齊y=0的位置
ax.spines['left'].set_position(('data',0)) #讓y軸對齊x=0的位置

plt.show()
```
