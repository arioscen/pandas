```python
import matplotlib.pyplot as plt
 
data = [2874, 1894, 1485, 1226, 1119, 985, 372, 92, 56, 8]
labels = ['Java', 'C#', 'JavaScript', 'PHP', 'C', 'C++', 'JPython', 'Ruby', 'R', 'Go']
plt.bar(range(len(data)), data, tick_label=labels, width=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

data = [2874, 1894, 1485, 1226, 1119, 985, 372, 92, 56, 8]
data.reverse()

labels = ['Java', 'C#', 'JavaScript', 'PHP', 'C', 'C++', 'JPython', 'Ruby', 'R', 'Go']
labels.reverse()

plt.barh(range(len(data)), data, tick_label=labels, height=0.5)

ax=plt.gca()

for label in ax.get_xticklabels()+ax.get_yticklabels():
    label.set_fontsize(16)

plt.show()
```
