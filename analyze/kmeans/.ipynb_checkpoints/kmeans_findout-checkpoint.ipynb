```python
import csv
with open('kmeans_class.csv','r') as f:
    d1 = 0
    d2 = 0
    d3 = 0
    date = ""
    for k in csv.reader(f):
        d1 = d2
        d2 = d3
        d3 = int(k[1])
        if d1 == d3 and d1 != d2 and d2 in [1,3,4,5]:
            print(date+","+str(d2))
        date = k[0]
```
