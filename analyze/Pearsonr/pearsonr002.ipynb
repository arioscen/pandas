```python
# 導入類別目錄
import csv
tags = []
with open('tag_distinct.csv', 'r') as f:
    for line in f:
        tags.append(line.strip())
tags
```

```python
# 設定時間區段
from datetime import datetime, timedelta
start = "2013-12-31"
end = "2017-06-15"
s = datetime.strptime(start, "%Y-%m-%d")
e = datetime.strptime(end, "%Y-%m-%d")
dates = []
while s <= e:
    sf = s.strftime("%Y-%m-%d")
    dates.append(sf)
    s += timedelta(1)
```

```python
# 產生dataframe
# 設為0.0001是為了避免pearsonr運算時警告
import pandas as pd
df = pd.DataFrame([list([0.0001]*len(tags))]*len(dates),index=dates,columns=tags)
df
```

```python
# 輸入每日銷售比率(rate=銷量(日)/銷量(年))
with open('tag_date_rate.csv', 'r') as f:
    for line in csv.reader(f):
        try:
            df.set_value(line[1],line[0],float(line[2]))
        except ValueError:
            df.set_value(line[1],line[0],0)
df
```

```python
# 去除不足一年的數據(2013、2017)
df.index[1097]
```

```python
df = df.drop(df.index[1097:])
df = df.drop(df.index[0])
df
```

```python
import pandas as pd
pm = pd.DataFrame([list([0.0000]*len(tags))]*len(tags),index=tags,columns=tags)
pv = pd.DataFrame([list([0.0000]*len(tags))]*len(tags),index=tags,columns=tags)
```

```python
from scipy.stats.stats import pearsonr
for i in range(0,len(df.columns)):
    i_name = df.columns[i]
    i_list = df.iloc[:,i]
    for j in range(0,len(df.columns)):
        j_name = df.columns[j]
        j_list = df.iloc[:,j]
        pear = pearsonr(i_list,j_list)
        pm_value = round(pear[0],4)
        pv_value = round(pear[1],4)
        pm.set_value(i_name,j_name,pm_value)
        pv.set_value(i_name,j_name,pv_value)
```

```python
# 找商品的相關係數
for i in range(0,len(pm.index)):
    for j in range(0,len(pm.columns)):
        pm_value = pm.iloc[i,j]
        if pm_value > 0.8 and pm_value < 0.9:
#         if pm_value > 0.6 and pm_value < 0.7:
            print("%s-%s-%.4f" % (pm.index[i],pm.columns[j],pm_value))
    
```

```python
from collections import Counter

documents = []

# find every tag's relation
for i in range(0,len(pm.index)):
    document = {}
    
    tag_row = pm.index[i]
    document['tag'] = tag_row
    
    counter = Counter()
    for j in range(0,len(pm.columns)):
        tag_column = pm.columns[j]
        tag_value = pm.iloc[i,j]
        if tag_column != tag_row:
            counter[tag_column] = tag_value
    
    # select top 10 relative tag
    tops = counter.most_common(10)
    
    for top in tops:
        document[top[0]] = top[1]
    documents.append(document)        
```

```python
import json
with open('correlation.json','w') as f:
    for document in documents:
        f.write(json.dumps(document,ensure_ascii=False)+"\n")
        
# mongoimport --db analyze --collection correlation --file weng/correlation.json
```
