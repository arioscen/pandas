```python
import csv
tags = []
with open('tag_distinct.csv', 'r') as f:
    for line in f:
        tags.append(line.strip())
```

```python
from datetime import datetime, timedelta
start = "2013-12-31"
end = "2017-06-15"
s = datetime.strptime(start, "%Y-%m-%d")
e = datetime.strptime(end, "%Y-%m-%d")
dates = []
while s <= e:
    sf = s.strftime("%Y-%m-%d")
    dates.append(sf)
    s += timedelta(1)
```

```python
import pandas as pd

df = pd.DataFrame([list([0]*len(tags))],index=dates,columns=tags)
df
```

```python
with open('tag_date_qty.csv', 'r') as f:
    for line in csv.reader(f):
        try:
            if int(line[2]) >= 0:
                df.set_value(line[1],line[0],line[2])
        except ValueError:
            df.set_value(line[1],line[0],0)
df
```

```python
macd = pd.DataFrame([list([0.0000]*len(tags))],index=dates,columns=tags)
for j in range(len(df.columns)):
    tag = df.iloc[:,j]
    ewma12 = pd.ewma(tag,span=1)
    ewma26 = pd.ewma(tag,span=100)
    dif = ewma12 - ewma26
    ewma9 = pd.ewma(dif,span=2)
    osc = dif - ewma9
    macd.iloc[:,j] = osc
    
macd
```

```python
macd_slope = pd.DataFrame([list([0.0000]*len(tags))],index=dates,columns=tags)
qty_slope = pd.DataFrame([list([0.0000]*len(tags))],index=dates,columns=tags)
for j in range(len(df.columns)):
    for i in range(2,len(df.index)): # 至少需要三天，所以從 2 開始
        if df.iloc[i-1,j] == 0:
            q_slope = df.iloc[i,j]-1
            qty_slope.set_value(qty_slope.index[i],qty_slope.columns[j],q_slope)
        else:
            q_slope = (df.iloc[i,j]-df.iloc[i-1,j])/df.iloc[i-1,j] # 今天的減掉昨天的
            qty_slope.set_value(qty_slope.index[i],qty_slope.columns[j],q_slope)

        m_slope = (macd.iloc[i-1,j]-macd.iloc[i-2,j])/macd.iloc[i-2,j] # macd數值是用來預測明天的，所以多減掉一天
        macd_slope.set_value(macd_slope.index[i],macd_slope.columns[j],m_slope)
```

```python
macd_slope = macd_slope.drop(macd_slope.index[:78])
macd_slope
```

```python
qty_slope = qty_slope.drop(qty_slope.index[:78])
qty_slope
```

```python
from scipy.stats.stats import pearsonr
pearsonr(macd_slope.iloc[:,100],qty_slope.iloc[:,100])
```
