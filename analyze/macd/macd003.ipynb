```python
import csv
tags = []
with open('tag_distinct.csv', 'r') as f:
    for line in f:
        tags.append(line.strip())
```

```python
from datetime import datetime, timedelta
start = "2013-12-31"
end = "2017-06-15"
s = datetime.strptime(start, "%Y-%m-%d")
e = datetime.strptime(end, "%Y-%m-%d")
dates = []
while s <= e:
    sf = s.strftime("%Y-%m-%d")
    dates.append(sf)
    s += timedelta(1)
```

```python
import pandas as pd

df = pd.DataFrame([list([0]*len(tags))],index=dates,columns=tags)
df
```

```python
with open('tag_date_qty.csv', 'r') as f:
    for line in csv.reader(f):
        try:
            if int(line[2]) >= 0:
                df.set_value(line[1],line[0],line[2])
        except ValueError:
            df.set_value(line[1],line[0],0)
df
```

```python
df_dif = pd.DataFrame([list([0.0000]*len(tags))],index=dates,columns=tags)
df_dem = pd.DataFrame([list([0.0000]*len(tags))],index=dates,columns=tags)
df_osc = pd.DataFrame([list([0.0000]*len(tags))],index=dates,columns=tags)
for j in range(len(df.columns)):
    tag = df.iloc[:,j]
    ewma12 = pd.ewma(tag,span=36)
    ewma26 = pd.ewma(tag,span=78)
    dif = ewma12 - ewma26
    # 加入dif的table
    df_dif.iloc[:,j] = dif
    
    ewma9 = pd.ewma(dif,span=27)
    # 加入dem的table
    df_dem.iloc[:,j] = ewma9
    
    osc = dif - ewma9
    #加入osc的table
    df_osc.iloc[:,j] = osc
```

```python
df_dif = df_dif.drop(df_dif.index[:78])
df_dem = df_dem.drop(df_dem.index[:78])
df_osc = df_osc.drop(df_osc.index[:78])
```

```python
import json
with open('df_dif.json','w') as f:
    for c in range(0,len(df_dif.columns)):
        data = {}
        tag = df_dif.columns[c]
        data['tag'] = tag+"_dif"
        for r in range(0,len(df_dif.index)):
            key = df_dif.index[r]
            value = round(df_dif.iloc[r,c],4)
            data[key] = value
        f.write(json.dumps(data,ensure_ascii=False)+"\n")
```
